```python
# =========================
# 02_feature_engineering.ipynb
# =========================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

behavior = pd.read_csv("../dataset/processed/behavior_features.csv")
preference = pd.read_csv("../dataset/processed/preference_features.csv")
lifecycle = pd.read_csv("../dataset/processed/lifecycle_features.csv")
```

```python
behavior.describe()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=behavior, x="total_orders")
plt.title("Total Orders Distribution")
plt.xlabel("Total Orders per User")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=behavior, x="total_products")
plt.title("Total Products Purchased per User")
plt.xlabel("Total Products")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(behavior["reorder_ratio"], bins=30)
plt.title("Reorder Ratio Distribution")
plt.xlabel("Reorder Ratio")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(behavior["avg_days_between_orders"], bins=30)
plt.title("Average Days Between Orders")
plt.xlabel("Days")
plt.show()
```
